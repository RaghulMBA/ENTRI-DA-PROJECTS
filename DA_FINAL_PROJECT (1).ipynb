The target variable Exited indicates whether a customer stayed with the bank or churned.

Identifies reasons behind customer churn by analyzing patterns such as low credit scores, inactive membership, high complaints, or low satisfaction.

**Prescriptive Analysis:**

Provides recommendations for the bank to improve retention, such as loyalty programs, targeted offers, and personalized customer support.

# Stage 2 – Data Cleaning and Pre-processing

Handle missing values (impute or drop)

Handle duplicates

Treat outliers if required

Check skewness and apply transformations

Convert data types if needed

Feature transformations (date parts, derived fields if required for analysis)

```python
# To Handle the missing values using (impute or drop)

df_dropna = df.dropna()

print("Original Shape:", df.shape)
print("After Dropping Missing Rows:", df_dropna.shape)
```

```python
# To Handle duplicates using drop function

duplicate_count = df.duplicated().sum()
print("Duplicate rows in the dataset:", duplicate_count)
```

```python
# To Treat outliers if required

import pandas as pd

Q1 = df['Age'].quantile(0.25)
Q3 = df['Age'].quantile(0.75)
IQR = Q3 - Q1

outliers = df[(df['Age'] < Q1 - 1.5*IQR) | (df['Age'] > Q3 + 1.5*IQR)]
print("Outliers before treatment:")
print(outliers)

df.describe()
```

```python
# To view Skewness & Transformations in these dataset

skew_vals = df.skew(numeric_only=True)
print("Skewness:\n", skew_vals)

for col in df.select_dtypes(include=[np.number]).columns:
    if abs(skew_vals[col]) > 1:
        df[col] = np.log1p(df[col])
```

```python
# To Convert data types if needed

# 1. Convert categorical columns

categorical_cols = ["Geography", "Gender", "Card Type"]
for col in categorical_cols:
    df[col] = df[col].astype("category")

# 2. Ensure numeric columns are numeric

numeric_cols = [col for col in df.columns if col not in categorical_cols]
df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric, errors="coerce")

print(df.dtypes)
```

```python
# To Using Feature transformations in these dataset

# 1. Age Groups
df["AgeGroup"] = pd.cut(df["Age"],
                        bins=[0, 25, 35, 45, 55, 65, 100],
                        labels=["<25", "25-35", "35-45", "45-55", "55-65", "65+"],
                        right=False)

# 2. Balance-to-Salary Ratio
df["BalanceSalaryRatio"] = (df["Balance"] / df["EstimatedSalary"]).replace([float("inf"), -float("inf")], 0).fillna(0)

# 3. Credit Score Category
df["CreditCategory"] = pd.cut(df["CreditScore"],
                              bins=[0, 580, 670, 740, 800, 900],
                              labels=["Poor", "Fair", "Good", "Very Good", "Excellent"],
                              right=False)

print(df[["Age", "AgeGroup","CreditScore", "CreditCategory","Balance", "EstimatedSalary"]].head(10))
```

```python
# Feature transformations to see the Card Type by its code

df["Card_Type_Code"] = df["Card Type"].map({
    "SILVER": 1,
    "GOLD": 2,
    "PLATINUM": 3,
    "DIAMOND": 4
})

# View the first 5 rows with the new column
df[["CustomerId","Card Type","Card_Type_Code", "EstimatedSalary"]].tail(10)
```
